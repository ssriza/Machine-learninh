#Task 2: Trip Pattern & Demand Analysis(25 marks)

```python
# Extract relevant datetime features from 'pickup_datetime' for modeling

df["pickup_date"] = df["pickup_datetime"].dt.date         # just date (YYYY-MM-DD)
df["pickup_hour"] = df["pickup_datetime"].dt.hour         # hour (0-23)
df["pickup_weekday"] = df["pickup_datetime"].dt.day_name()  # weekday name
```

```python
#Analyze trip distribution over time to identify trends such as daily and hourly peak periods, weekday.


# 1️) Daily trip distribution
daily_trips = (
    df.groupby("pickup_date")
      .size()                       # counts rows per group
      .reset_index(name="total_trips")  # rename the count column
      .sort_values("pickup_date")   # sort by date
)

# 2️) Hourly trip distribution
hourly_trips = (
    df.groupby("pickup_hour")
      .size()
      .reset_index(name="total_trips")
      .sort_values("pickup_hour")
)

# 3️) Weekday trip distribution
weekday_trips = (
    df.groupby("pickup_weekday")
      .size()
      .reset_index(name="total_trips")
      .sort_values("pickup_weekday")
)


# Show results
print(f'Daily Trips: {daily_trips}')
print(f'Hourly Trips: {hourly_trips}')
print(f'Weekday Trips: {weekday_trips}')
```

```python
# Visualization for daily trips, hourly, and weekday trips

# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(20,6))

# 1️. Daily trips (line plot)
sns.lineplot(data=daily_trips, x="pickup_date", y="total_trips", ax=axes[0])
axes[0].set_title("Daily Trip Distribution")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Number of Trips")
axes[0].tick_params(axis='x', rotation=45)

# 2️.  Hourly trips (bar plot)
sns.barplot(data=hourly_trips, x="pickup_hour", y="total_trips", color="skyblue", ax=axes[1])
axes[1].set_title("Hourly Trip Distribution")
axes[1].set_xlabel("Hour of Day")
axes[1].set_ylabel("Number of Trips")

# 3️. Weekday trips (bar plot)
sns.barplot(data=weekday_trips, x="pickup_weekday", y="total_trips", palette="viridis", ax=axes[2])
axes[2].set_title("Weekday Trip Distribution")
axes[2].set_xlabel("Weekday")
axes[2].set_ylabel("Number of Trips")
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Load TLC Zone Lookup (Borough info)
# -----------------------------------------------------

zone_lookup = pd.read_csv("taxi_zone_lookup.csv")
zone_lookup
```

```python
#Compute total revenue, average fare per trip, and analyze fare distribution across different boroughcities


# Rename for consistency
zone_lookup = zone_lookup.rename(columns={"LocationID":"pulocation_id", "Borough":"pickup_borough"})

# -----------------------------------------------------
# merge trip data with borough lookup
# -----------------------------------------------------
df_with_borough = df.merge(zone_lookup, on="pulocation_id")


```

```python
# -----------------------------------------------------
# 1. Total revenue
# -----------------------------------------------------
total_revenue = df_with_borough["total_amount"].sum()

# -----------------------------------------------------
# 2. Average fare per trip
# -----------------------------------------------------
avgerage_fare = df_with_borough["fare_amount"].sum()

# -----------------------------------------------------
# 3. Fare distribution by borough
# -----------------------------------------------------
fare_by_borough = (
    df_with_borough.groupby("pickup_borough")
    .agg(
        trip_count=("pickup_borough", "size"),           # count of trips
        total_revenue=("total_amount", "sum"),           # sum of total_amount
        avg_fare=("fare_amount", "mean")                 # average fare_amount
    )
    .sort_values("total_revenue", ascending=False)      # sort by total revenue descending
    .reset_index()                                     # optional: make pickup_borough a column
)
```

```python
# Show results
print(total_revenue)
print(avgerage_fare)
print(fare_by_borough)
```

```python
#  Boxplot to see fare distribution per borough
plt.figure(figsize=(12,6))
sns.boxplot(data=df_with_borough, x="pickup_borough", y="fare_amount", palette="Set2")
plt.title("Fare Distribution by Borough")
plt.xlabel("Borough/City")
plt.ylabel("Fare Amount ($)")
plt.xticks(rotation=45)
plt.ylim(0, 200)  # optional: limit outliers for better visualization
plt.tight_layout()
plt.show()
```

```python
# Barplot of average fare per borough
plt.figure(figsize=(10,5))
sns.barplot(data=fare_by_borough, x="pickup_borough", y="avg_fare", palette="Blues_d")
plt.title("Average Fare by Borough")
plt.xlabel("Borough/City")
plt.ylabel("Average Fare ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
